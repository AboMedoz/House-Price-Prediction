```python
import joblib
import os
import random

import pandas as pd
```

```python
ROOT = os.path.dirname(os.getcwd())
DATA_PATH = os.path.join(ROOT, 'data', 'processed')
MODELS_PATH = os.path.join(ROOT, 'models')
```

```python
model = joblib.load(os.path.join(MODELS_PATH, 'random_forest.pkl'))
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'processed.csv'))
```

```python
rand_row = random.randint(0, df.shape[0])
```

```python
x = df.drop(columns=['2025-01-31'])
y = df['2025-01-31']
```

```python
prediction = model.predict(x.iloc[[rand_row]])[0]
gt = y.iloc[rand_row]
```

```python
print(f"Predicted : ${prediction:,.2f}")
print(f"Actual    : ${gt:,.2f}")
print(f"Difference: ${abs(prediction - gt):,.2f}")
```
